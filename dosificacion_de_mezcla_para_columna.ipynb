# *Dosificación de Mezcla para Columnas*

*Planteamiento:*

Un ingeniero residente necesita calcular cuántos sacos de cemento se requieren para vaciar una columna de sección circular.

Se sabe que para un concreto estándar de 210 kg/cm2 (3000 psi), se consumen 9.5 sacos de cemento por cada metro cúbico de mezcla.

*Requerimientos:*

1. Solicitar la altura de la columna en metros.
2. Solicitar el diámetro de la columna en metros.
3. Calcular el volumen de la columna: $$v= \pi * r^2 * h$$
. (Sugerencia: usar 3.1416
 o importar math.pi).
4. Calcular la cantidad de sacos:
$$Sacos=v * 9.5$$

*Prueba del programa:*

Para una columna de 3 metros de alto y 0.4 metros de diámetro (r=0.2
):

Volumen ≈0.377 m3
.
Sacos necesarios ≈3.58
 (el estudiante verá aquí la importancia de los tipos decimales).

````python
 #[ENTRADA]: Solicitar la altura de la columna en metros
 # Solicitar el diámetro de la columna en metros.
 h = float(input("Ingrese la altura de la columna en metros: "))
 d = float(input("Ingrese el diámetro de la columna en metros: "))
 # PROCESO: valcula el radio, calcula el volumen de la columna, calcula sacos
 r = d/2
 v = 3.14 * r**2 * h
 sacos = v * 9.5
 # SALIDA: imprime el resultado
 print("Volumen ≈", v, "m3")
 print("Sacos necesarios ≈", sacos, "m3")
#
````

```markdown
## Explicación del Código de Cálculo de Sacos de Cemento

Este código realiza el cálculo de la cantidad de sacos de cemento necesarios para una columna de sección circular, basándose en la altura y el diámetro de la columna.

```python
h = float(input("Ingrese la altura de la columna en metros: "))
```
- **`h = float(input("Ingrese la altura de la columna en metros: "))`**:
  - `input("...")`: Muestra el mensaje "Ingrese la altura de la columna en metros: " al usuario en la consola y espera a que el usuario ingrese un valor.
  - `float(...)`: Convierte la entrada del usuario (que siempre es una cadena de texto) a un número de coma flotante (decimal). Esto es necesario porque la altura puede ser un valor decimal.
  - `h = ...`: Asigna el valor numérico ingresado por el usuario a la variable `h` (altura).

```python
d = float(input("Ingrese el diámetro de la columna en metros: "))
```
- **`d = float(input("Ingrese el diámetro de la columna en metros: "))`**:
  - Similar a la línea anterior, solicita al usuario el diámetro de la columna en metros.
  - `float(...)`: Convierte la entrada a un número de coma flotante.
  - `d = ...`: Asigna el valor numérico a la variable `d` (diámetro).

```python
r = d/2
```
- **`r = d/2`**:
  - Calcula el radio (`r`) de la columna dividiendo el diámetro (`d`) por 2. El radio es fundamental para el cálculo del volumen de un círculo.

```python
v = 3.14 * r**2 * h
```
- **`v = 3.14 * r**2 * h`**:
  - Calcula el volumen (`v`) de la columna. Esta es la fórmula para el volumen de un cilindro (`π * radio² * altura`).
  - `3.14`: Es una aproximación del valor de Pi (π).
  - `r**2`: Eleva el radio (`r`) al cuadrado.
  - `*`: Son operadores de multiplicación.

```python
sacos = v * 9.5
```
- **`sacos = v * 9.5`**:
  - Calcula la cantidad de sacos de cemento (`sacos`) necesarios.
  - `v`: Es el volumen de la columna calculado previamente.
  - `9.5`: Es el consumo de sacos de cemento por cada metro cúbico de mezcla, según el planteamiento del problema.

```python
print("Volumen ≈", v, "m3")
```
- **`print("Volumen ≈", v, "m3")`**:
  - Muestra el resultado del volumen de la columna al usuario en la consola.
  - "Volumen ≈": Es un texto fijo que se imprime.
  - `v`: Es el valor de la variable `v` (volumen).
  - "m3": Es un texto fijo que indica las unidades (metros cúbicos).

```python
print("Sacos necesarios ≈", sacos, "m3")
```
- **`print("Sacos necesarios ≈", sacos, "m3")`**:
  - Muestra el resultado de la cantidad de sacos de cemento necesarios al usuario en la consola.
  - "Sacos necesarios ≈": Es un texto fijo que se imprime.
  - `sacos`: Es el valor de la variable `sacos`.
  - "m3": Es un texto fijo que indica las unidades (metros cúbicos), aunque en este caso sería más apropiado "sacos" o dejarlo sin unidad explícita si se entiende el contexto.
