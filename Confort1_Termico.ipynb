# *Métodos y cálculos*

* Índice de temperatura y humedad (producción intensiva)

$$ITH = 0.72(T_{bs} + T_{bh}) + 40.6$$

Donde:


$T_{bs}$ Temperatura bulbo seco ($^oC$)


$T_{bh}$ Temperatura de bulbo húmedo ($^oC$)

Para la correlación de los datos de la especie bovina en la producción de leche se recomienda
utilizar los límites propuestos por National Weather Service:

Sí $ITH ≤ 70$ ambiente no estresante

Sí $ITH > 70 \space And \space ITH ≤ 78$ situación de alerta por estrés térmico

Sí $ITH > 78 \space And \space ITH ≤ 83$ peligro por estrés térmico

Sí $ITH > 83$ situación crítica

````python
# ENTRADA: Pedir tbs (Temperatura de bulbo secoo) y tbh (Temperatura de bulbo húmedo)
tbs = float(input("Ingrese la temperatura de bulbo seco, Grados Centigrados: "))
tbh = float(input("Ingrese la temperatura de bulbo húmedo, Grados Centigrados: "))

# PROCESO: Calculo de ith (índice de Temperatura y Humedad)
ith = 0.72 * (tbs + tbh) + 40.6

# SALIDA: Imprimir resultado de la evaluación del ith y el confort térmico
if ith <= 70:
  print("El índice de temperatura y humedad es: ", ith)
  print("Ambiente no estresante")

elif ith > 70 and ith <= 78:
  print("El índice de temperatura y humedad es: ", ith)
  print("Situación de alerta por estrés térmico")

elif ith > 78 and ith <= 83:
  print("El índice de temperatura y humedad es: ", ith)
  print("peligro por estrés térmico")

elif ith > 83:
  print("El índice de temperatura y humedad es: ", ith)
  print("Situación crítica")
````

## Teoría del Confort Térmico en Ganado Vacuno

El confort térmico en el ganado vacuno se refiere al estado en el que el animal no experimenta estrés por calor ni por frío, y su metabolismo no necesita gastar energía adicional para mantener una temperatura corporal interna estable. Este estado es crucial para la salud, bienestar y productividad del ganado, incluyendo la producción de leche, carne y la reproducción.

### Factores que Afectan el Confort Térmico:

1.  **Temperatura del aire (Temperatura de bulbo seco - Tbs):** Es el factor más obvio y directo. Temperaturas elevadas o muy bajas pueden causar estrés.
2.  **Humedad relativa (Humedad de bulbo húmedo - Tbh):** La humedad dificulta la disipación del calor corporal a través de la evaporación del sudor, aumentando la sensación térmica en ambientes cálidos.
3.  **Radiación solar:** La exposición directa al sol incrementa la carga térmica sobre el animal.
4.  **Velocidad del viento:** Un viento suave puede ayudar a la disipación de calor en ambientes cálidos, pero un viento fuerte en climas fríos puede aumentar la sensación de frío.
5.  **Precipitación:** La lluvia puede mojar el pelaje, reduciendo su capacidad aislante en frío o contribuyendo al enfriamiento evaporativo en calor.
6.  **Características del animal:** La raza (ej. Bos taurus vs. Bos indicus), edad, nivel de producción, estado fisiológico (gestación, lactancia) y color del pelaje influyen en su tolerancia a las condiciones térmicas.

### Zona de Confort Térmico (ZCT) o Zona de Termoneutralidad:

Es el rango de temperaturas ambientales dentro del cual un animal puede mantener su temperatura corporal interna constante sin aumentar su tasa metabólica. Por debajo de este rango, el animal gasta energía para producir calor (termogénesis); por encima, gasta energía para disipar calor (termólisis).

### Estrés por Calor:

Ocurre cuando la temperatura ambiente y/o la humedad son demasiado altas y el animal no puede disipar el exceso de calor corporal de manera eficiente. Síntomas incluyen:
*   Aumento de la frecuencia respiratoria (jadeo).
*   Reducción del consumo de alimento.
*   Disminución de la producción de leche o carne.
*   Deterioro de la eficiencia reproductiva.
*   Aumento del consumo de agua.
*   Búsqueda de sombra.

### Estrés por Frío:

Se produce cuando la temperatura ambiente es demasiado baja y el animal debe aumentar su producción de calor metabólico para mantener su temperatura corporal. Síntomas incluyen:
*   Piloerección (erizar el pelo).
*   Temblor.
*   Aumento del consumo de alimento (para generar calor).
*   Disminución de la producción.

### Índices de Confort Térmico:

Son herramientas que combinan múltiples variables ambientales (principalmente temperatura y humedad) en un solo valor para cuantificar el nivel de estrés térmico. El Índice de Temperatura y Humedad (ITH) es uno de los más utilizados, como se describe en la metodología, y permite clasificar el ambiente en diferentes categorías de estrés para tomar decisiones de manejo. Estos índices son vitales para sistemas de alerta temprana y la implementación de estrategias de adaptación al cambio climático en la ganadería.

### Explicación del Código de Confort Térmico

El siguiente código Python calcula el Índice de Temperatura y Humedad (ITH) basándose en la temperatura de bulbo seco (Tbs) y la temperatura de bulbo húmedo (Tbh), y luego clasifica el nivel de confort térmico para el ganado vacuno según el valor del ITH.
